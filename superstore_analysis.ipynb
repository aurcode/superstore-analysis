```python
import matplotlib.pyplot as plt
import numpy as np, pandas as pd
#from scipy import stat
```

```python
df_superstore = pd.read_csv('https://raw.githubusercontent.com/aurcode/superstore-analysis/main/sample-superstore.csv')
print('df_superstore Data set have a dimension of ', df_superstore.shape, ' records and columns respectively')
```

```python
print('Here we can see some basic info from the data set')
print(' ')
print('df_superstore')
print(' ')
df_superstore.info()
```

# ¿Cuál es la categoría de productos más vendida?

```python
best_selling_category = df_superstore.groupby(['Category']).agg({'Sales':'sum', 'Profit':'sum'})
best_selling_category
```

```python
best_selling_category.plot.bar()
```

# ¿Quién es el cliente que compra más?

```python
customers_sales = df_superstore.groupby(['Customer Name']).agg({'Sales':'sum'})
customers_who_buy_more = customers_sales.sort_values(['Sales'],ascending=False).head(10)
customers_who_buy_more
```

```python
customers_who_buy_more.plot.barh()
```

# En que region se hacen mas ventas?

```python
region_sales = df_superstore.groupby(['Region']).agg({'Sales':'count'})
region_sales
```
